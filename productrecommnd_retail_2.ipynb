## Create the Basket Data while Using The Transaction From West Region

Seperti yang kta lihat dari visualisasi di atas, sebagian besar transaksi berasal dari West Region. Jadi, untuk membuat proyek ini lebih sederhana dan lebih personal, mari kita batasi data yang akan kita gunakan hanya untuk transaksi yang berasal dari West Region. kami akan membuat data keranjang. Data keranjang ini akan berisi Kuantitas setiap item yang dibeli per transaksi. 

```python
basket_plus = (data[data['Region']=='West'].groupby(['Order ID', 'Product Name'])['Quantity']
               .sum().unstack().reset_index().fillna(0).set_index('Order ID'))
basket_plus
```

In market basket analysis, jumlah setiap barang yang dibeli tidak terlalu penting. Yang penting adalah apakah suatu barang dibeli atau tidak. Karena, kami hanya ingin tahu, apa hubungan membeli beberapa item dan membeli beberapa lainnya. Jadi, kita perlu mengkodekan data keranjang menjadi data biner yang menunjukkan apakah suatu barang dibeli (1) atau tidak (0). 

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_encode_plus = basket_plus.applymap(encode_units)
basket_encode_plus
```

Dalam market basket analysis, kita akan mengungkap hubungan antara 2 atau lebih item yang dibeli menurut data historis. Jadi, kurang bermanfaat jika transaksi hanya membeli satu item saja. Maksud saya, bagaimana kita bisa menemukan hubungan antara 2 item atau lebih jika hanya ada 1 item yang dibeli? Oleh karena itu, langkah selanjutnya adalah menyaring transaksi yang membeli lebih dari 1 item. Inilah cara melakukannya.


```python
basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

Setelah menghasilkan dataset di atas, sekarang saatnya kita menggunakan algoritma apriori. Algoritma Apriori hanya digunakan untuk menemukan item yang sering dibeli di dataset. 

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.002, 
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

Dalam penerapan algoritma apriori, kita dapat mendefinisikan data frequent yang kita inginkan dengan memberikan nilai support. Dalam hal ini saya mendefinisikan barang yang sering dibeli sebagai barang yang dibeli sebanyak 0,2% dari keseluruhan transaksi, artinya saya akan memberikan nilai support 0,002. Setelah itu, saya menambahkan kolom lain yang disebut panjang yang berisi jumlah barang yang dibeli.

```python
frequent_itemsets_plus[ (frequent_itemsets_plus['length']  == 1) &
                        (frequent_itemsets_plus['support'] >= 0.002) ]
```

Setelah menerapkan algoritma apriori dan menemukan barang yang sering dibeli, kini saatnya kita menerapkan aturan asosiasi. Dari aturan asosiasi, kami dapat mengekstrak informasi dan bahkan menemukan pengetahuan tentang item mana yang lebih efektif untuk dijual bersama. Itulah inti dari proyek ini.
